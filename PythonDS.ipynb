```python
import os
import requests
import pandas as pd
from tqdm.notebook import tqdm
from newspaper import Article
from datetime import datetime, timedelta

# -----------------------
# 1Ô∏è‚É£ Param√®tres de base
# -----------------------

TICKERS = {
    "MSFT": "Microsoft",
    "NVDA": "Nvidia",
    "AAPL": "Apple",
    "AMZN": "Amazon",
    "GOOGL": "Alphabet (Class A)",
    "GOOG": "Alphabet (Class C)",
    "META": "Meta Platforms",
    "TSLA": "Tesla",
    "BRK-B": "Berkshire Hathaway (Class B)",
    "AVGO": "Broadcom",
    "JPM": "JPMorgan Chase",
    "LLY": "Eli Lilly",
    "V": "Visa",
    "MA": "Mastercard",
    "UNH": "UnitedHealth Group",
    "XOM": "Exxon Mobil",
    "WMT": "Walmart",
    "COST": "Costco Wholesale",
    "HD": "Home Depot",
    "NFLX": "Netflix",
}

START_DATE = datetime(2024, 1, 1)
END_DATE = datetime(2024, 12, 31)
OUTPUT_DIR = "news_articles"
os.makedirs(OUTPUT_DIR, exist_ok=True)

# -----------------------
# 2Ô∏è‚É£ Fonction : GDELT API
# -----------------------

def get_articles_from_gdelt(company_name, start_date, end_date):
    """T√©l√©charge les m√©tadonn√©es d'articles GDELT contenant le nom de la soci√©t√©."""
    base_url = "https://api.gdeltproject.org/api/v2/doc/doc"
    results = []

    # On d√©coupe la p√©riode par mois pour √©viter les requ√™tes trop lourdes
    current = start_date
    while current <= end_date:
        next_month = (current + timedelta(days=32)).replace(day=1)
        params = {
            "query": company_name,
            "mode": "ArtList",
            "format": "json",
            "maxrecords": "250",
            "startdatetime": current.strftime("%Y%m%d000000"),
            "enddatetime": min(next_month, end_date).strftime("%Y%m%d235959"),
        }

        try:
            r = requests.get(base_url, params=params, timeout=15)
            if r.status_code == 200 and r.text.strip():
                data = r.json()
                if "articles" in data:
                    results.extend(data["articles"])
        except Exception as e:
            print(f"‚ö†Ô∏è Erreur GDELT ({company_name}, {current.strftime('%Y-%m')}) : {e}")

        current = next_month

    if not results:
        return pd.DataFrame()

    df = pd.DataFrame(results)
    keep_cols = ["title", "seendate", "sourceCountry", "url", "domain"]
    df = df[[c for c in keep_cols if c in df.columns]]
    df["date"] = pd.to_datetime(df["seendate"], errors="coerce").dt.date
    return df.dropna(subset=["url", "title"])

# -----------------------
# 3Ô∏è‚É£ Fonction : Texte complet
# -----------------------

def download_article_text(url, timeout=10):
    """R√©cup√®re le texte int√©gral de l'article avec Newspaper3k."""
    try:
        article = Article(url)
        article.download()
        article.parse()
        text = article.text.strip()
        return text if len(text) > 200 else None  # √©vite les articles trop courts
    except Exception as e:
        # Erreurs r√©seau, parse, timeout, etc.
        return None

# -----------------------
# 4Ô∏è‚É£ Boucle principale
# -----------------------

for ticker, company in tqdm(TICKERS.items(), desc="üì° Scraping GDELT 2024", position=0):
    try:
        df = get_articles_from_gdelt(company, START_DATE, END_DATE)
        if df.empty:
            print(f"‚ùå Aucun article trouv√© pour {company}")
            continue

        df["Ticker"] = ticker
        df["Company"] = company

        texts = []
        for url in tqdm(df["url"], desc=f"üì∞ T√©l√©chargement {ticker}", position=1, leave=False):
            text = download_article_text(url, timeout=10)
            texts.append(text)

        df["Content"] = texts
        df = df.dropna(subset=["Content"])
        df.to_csv(f"{OUTPUT_DIR}/{ticker}_2024.csv", index=False)

        print(f" {ticker} : {len(df)} articles enregistr√©s")

    except Exception as e:
        print(f" Erreur inattendue sur {ticker} ({company}) : {e}")

print(" Scraping termin√© !")
```

```python
import pandas as pd
from transformers import BertTokenizer, BertForSequenceClassification, pipeline
import re
from typing import Dict, List

# ==============================================================================
# 1. CHARGEMENT DU MOD√àLE FINBERT & OUTILS
# ==============================================================================

# Le mod√®le et le tokenizer sont charg√©s une seule fois
try:
    finbert = BertForSequenceClassification.from_pretrained('yiyanghkust/finbert-tone', num_labels=3)
    tokenizer = BertTokenizer.from_pretrained('yiyanghkust/finbert-tone')
    # Cr√©ation du pipeline avec troncature activ√©e par d√©faut (pour g√©rer les longs textes)
    nlp_pipeline = pipeline(
        "sentiment-analysis",
        model=finbert,
        tokenizer=tokenizer,
        truncation=True  # Assure que FinBERT peut g√©rer les articles longs
    )
    print("‚úÖ Mod√®le FinBERT charg√©.")
except ImportError:
    print("‚ö†Ô∏è ERREUR: La librairie 'transformers' n'est pas install√©e.")
    nlp_pipeline = None

# ==============================================================================
# 2. FONCTION DE QUANTIFICATION [-1, +1]
# ==============================================================================

def quantify_polarite(text: str) -> float:
    """
    Analyse le sentiment d'un texte et le convertit en score continu [-1, +1].
    S'assure que le texte est trait√©, qu'il soit long ou court.
    """
    if not nlp_pipeline or not text:
        return 0.0

    try:
        # Ex√©cution du mod√®le (truncation=True est pass√© lors de la cr√©ation du pipeline)
        results = nlp_pipeline(text)
        resultat = results[0]

        label = resultat['label']
        score = resultat['score']

        # Quantification :
        if label == 'LABEL_1':  # Positif
            return score
        elif label == 'LABEL_2':  # N√©gatif
            return -score
        else:  # Neutre (LABEL_0)
            return 0.0

    except Exception as e:
        print(f"Erreur de pipeline pour un article: {e}")
        return 0.0

# ==============================================================================
# 3. APPLICATION AU DATAFRAME
# ==============================================================================

# Simulation du DataFrame (Exemple de vos donn√©es apr√®s l'√©tape GDELT)
df_articles = pd.DataFrame({
    'article_id': [1, 2, 3],
    'ticker': ['MSFT', 'AAPL', 'TSLA'],
    'Article_Text': [
        "Microsoft just announced record-breaking quarterly profits, exceeding all analyst expectations.",
        "A small dip in iPhone sales in China was noted today, but the company's overall financial health remains sound and dividend payments are stable.",
        "The company is facing severe liquidity problems and must declare bankruptcy next week, analysts predict."
    ]
})

print("\n--- D√©but de l'Analyse de Sentiment ---")

# Appliquer la fonction directement √† la colonne de texte
# NOTE : Cette m√©thode donne un score GLOBAL √† l'article, pas un score cibl√©.
df_articles['Score_Final_Polarite'] = df_articles['Article_Text'].apply(quantify_polarite)

# Affichage des r√©sultats
print("\nR√©sultats de l'analyse de sentiment (Score: [-1 √† +1]) :")
print(df_articles[['article_id', 'ticker', 'Score_Final_Polarite']])
```
